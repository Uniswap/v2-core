```python
import numpy as np

import pandas
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('fee.txt', sep='\\s+', verbose=True)
df
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.plot(df['rFactor'] / (10 ** 18 ), (df['fee'] / (10 ** 16) - 0.3))
ax1.grid(True)
ax1.set_ylim(-0.2, 0.4)
ax1.set_xlim(0, 2)

plt.show()
```

```python
df = pandas.read_csv('feeIncrease.txt', sep='\\s+', verbose=True)
df
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.plot(df['block'], df['shorEMA'] / (10 ** 18 ))
ax1.plot(df['block'], df['longEMA'] / (10 ** 18 ))
ax1.plot(df['block'], df['rFactor'] / (10 ** 18 ))

ax1.set_xticks(np.arange(0, 10800, 900))
ax1.set_ylim(0, 2)
ax1.grid(True)

ax2 = ax1.twinx()
ax2.plot(df['block'], df['fee'] / (10 ** 18 ), color = 'r')
ax2.set_ylabel('Fee', color='r')
ax2.set_ylim(0, 0.006)
```
